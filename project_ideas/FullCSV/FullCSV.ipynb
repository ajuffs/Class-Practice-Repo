```python
import nltk
import pandas as pd
import numpy as np
```

```python
stream_df = pd.read_csv('Northernlion_Full.csv',index_col=None, header=0)
stream_df.drop('GamesPlayed',axis=1, inplace=True)
stream_df
```

```python
#Create Date Col as Month DD, YYYY
formatted = []
order = [1,0,2]
for date in stream_df['Stream start time']:
    dmy = date.split(' ')[1:-1] #Date/Month/Year
    dmy[0] = dmy[0][:-2] #Remove day suffixes
    mdy = [dmy[i] for i in order]
    formatted.append(str(mdy[0] + " " + mdy[1] + ", " + mdy[2]))
formatted[:15]
```
