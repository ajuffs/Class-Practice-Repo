## (4) Accessing routines

```python
# Method 1:
newfilter = songs["Year"] == 2015    # boolean series
newsongs = songs.loc[newfilter,:]
newsongs.head()
newsongs.tail()
newsongs.info()
```

```python
# Method 2:
newsongs = songs.iloc[-100:,:]
newsongs.head()
newsongs.tail()
```

```python
# Method 1 without defining the filter previously
oldsongs = songs.loc[songs["Year"] == 1965,:]
oldsongs.head()
oldsongs.tail()
#oldsongs.info()

# Songs from 2000 to 2015 (2000 to the end of the data set)
millenial = songs.loc[songs["Year"] >= 2000,:]
millenial.info()
```

```python
# Madonna Songs
len(songs.loc[songs["Artist"]=='madonna',:])
songs.loc[songs["Artist"]=='madonna',:]


# Madonna Songs Top 10
songs.loc[(songs["Artist"]=='madonna') & (songs["Rank"]<=10),:]


# Madonna Songs from 2000+

# DOES NOT WORK:
    #songs.loc[songs["Artist"]=='madonna' and songs["Year"]>=2000,:]
    
songs.loc[(songs["Artist"]=='madonna') & (songs["Year"]>=2000),:]
```


## (7) Filling in missing values
- We really need to take care of the missing lyric values. 
- What are good strategies? Considerations?

```python
# replace NaN (and instrumental) with empty string

songs.fillna('')    #returning new data frame
songs[:50]          #not changed

#songs = songs.fillna('')
#songs[:50]

songs.fillna('', inplace = True)
songs.head()
```

```python
# LIST COMPREHENSION
#[x for x in [1,2,3,4]]

toklengths = [len(x.split()) for x in foo['Lyrics']]
toklengths[:100]

# data frame column is a data frame series
songs["Length"] = pd.Series(toklengths)
songs.head()


#defining function ahead of time
def tokenize(txt):
    return txt.split()

songs['Tokens'] = songs["Lyrics"].map(tokenize)
songs.head()


#defining simple function and then throw away USING LAMBDA
    #use on the fly without giving a name
songs['tokens2'] = songs["Lyrics"].map(lambda x: x.split())
songs.head()

del songs["tokens2"]
songs.head()


def get_avg_len(toks):
    lens = [len(t) for t in toks]
    if lens(toks) == 0: return 0
    else: return sum(lens)/len(lens)
    
sent = "colorless ideas sleep furiously".split()
sent
# get_avg_len(sent)
#songs[avg_wd_len] = songs["Tokens"].map(get_avg_len)
```

## (8) A detour: text processing basics
- NLTK functions
- How to word-tokenize
- How to sentence-tokenize
- Frequency distribution
- Type-token ratio

```python
# SEE LAB13 for more nltk info 
import nltk
nltk.corpus.brown.words()
nltk.word_tokenize("Hello, world!")

#type token ratio TTR
    #measure for vocabulary richness
    #but as text length gets longer, variety does not increase linearly, and there is a ceiling
    #must compare 2 texts of equal lengths!!!!

rose = "rose is a rose is a rose is a rose".split()
rose
len(set(rose))

def get_ttr(toks):
    if len(toks) ==0: return 0
    else: return len(set(toks))/len(toks)

get_ttr(rose)

#0 division error unless and if statement to the function
songs['ttr']=songs['Tokens'].map(get_ttr)
songs.head()
```
